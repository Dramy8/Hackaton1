### Data Cleaning and Preprocessing

```python
df_clean = df.copy()

# Standardize text columns to lowercase
df_clean["interest"] = df_clean["interest"].str.lower()
df_clean["content_category"] = df_clean["content_category"].str.lower()
df_clean["preferred_format"] = df_clean["preferred_format"].str.lower()
df_clean["activity_type"] = df_clean["activity_type"].str.lower()
df_clean["content_format"] = df_clean["content_format"].str.lower()

# Fill missing categorical values
df_clean["preferred_format"] = df_clean["preferred_format"].fillna("unknown")
df_clean["recommended_content"] = df_clean["recommended_content"].fillna("Unknown recommendation")

# Fill missing numerical values with the median
df_clean["engagement_score"] = df_clean["engagement_score"].fillna(df_clean["engagement_score"].median())
df_clean["time_spent_minutes"] = df_clean["time_spent_minutes"].fillna(df_clean["time_spent_minutes"].median())
df_clean["clicks_count"] = df_clean["clicks_count"].fillna(df_clean["clicks_count"].median())
df_clean["user_rating"] = df_clean["user_rating"].fillna(df_clean["user_rating"].median())

# Check missing values after cleaning
df_clean.isnull().sum()
```

### Statistical Analysis with SciPy

#### Chi-Square Test: Interest and Activity Type

```python
interest_activity_table = pd.crosstab(df_clean["interest"], df_clean["activity_type"])

interest_activity_table
```

The cross-tabulation shows that viewed is the most common activity type across almost all interests. 
Technology has the highest number of viewed interactions, followed by music and gaming.

Purchased is the least frequent activity type overall, which is realistic because purchasing requires more commitment than simply viewing or liking content. 
Fitness also shows relatively high completed and saved activity counts, suggesting that users interested in fitness may be more likely to save or finish practical content.

```python
chi2, p_value, dof, expected = stats.chi2_contingency(interest_activity_table)

print("Chi-square statistic:", chi2)
print("P-value:", p_value)
print("Degrees of freedom:", dof)
```

The chi-square test gives a p-value of 0.2095. 
Since this value is greater than 0.05, we do not find a statistically significant relationship between user interest and activity type.

This means that, in this synthetic dataset, the type of activity does not strongly depend on the user's main interest. 
Although some differences appear in the cross-tabulation table, they are not statistically strong enough to confirm a relationship.

#### Pearson Correlation: Content Popularity and Engagement Score

```python
pearson_corr, pearson_p_value = stats.pearsonr(
    df_clean["content_popularity"],
    df_clean["engagement_score"]
)

print("Pearson correlation:", pearson_corr)
print("P-value:", pearson_p_value)
```

The Pearson correlation between content_popularity and engagement_score is 0.4623, which indicates a moderate positive relationship. 
This means that more popular content tends to be associated with higher user engagement.

The p-value is extremely small and below 0.05, so the relationship is statistically significant. 
This suggests that content popularity is meaningfully related to engagement in this synthetic dataset.

However, correlation does not prove causation. 
This result shows that popularity and engagement move together, but it does not prove that popularity directly causes higher engagement.

### Recommendation System

After exploring the data, I used the main patterns to build a simple recommendation logic. 
The idea is to give more points when the content matches the user's profile and behavior.

The original dataset already contains a recommendation_score column, which was part of the synthetic data generation process. 
In the recommendation system section, I create a new score called rule_based_score to show my own recommendation logic step by step.

#### Recommendation Scoring Logic

The recommendation score is based on simple rule-based logic. 
The goal is to simulate personalized recommendations without using a complex machine learning model.

The score increases when:
- the user's interest matches the content category
- the user's preferred format matches the content format
- the user performs a stronger activity, such as saving, completing, or purchasing
- the content is more popular

```python
activity_weights = {
    "viewed": 1,
    "liked": 2,
    "shared": 3,
    "saved": 4,
    "completed": 5,
    "purchased": 7
}

def calculate_recommendation_score(row):
    score = 10
    
    if row["interest"] == row["content_category"]:
        score += 5
        
    if row["preferred_format"] == row["content_format"]:
        score += 3
        
    score += activity_weights[row["activity_type"]]
    score += row["content_popularity"] / 20
    
    return score
```

This function creates a simple rule-based recommendation score. 
The score starts with a base value of 10, then increases when the content matches the user's interest, when the format matches the user's preferred format, when the user activity shows stronger commitment, and when the content is popular.

This approach simulates a simple AI-powered recommendation logic without using a complex machine learning model.

```python
df_clean["rule_based_score"] = df_clean.apply(calculate_recommendation_score, axis=1)

df_clean[["interest", "content_category", "preferred_format", "content_format", "activity_type", "content_popularity", "rule_based_score"]].head()
```

The apply() method is used to calculate a recommendation score for each row of the dataset. 
Since the score depends on several columns from the same row, axis=1 is used to apply the function row by row.

The result is stored in a new column called rule_based_score.

The rule_based_score column was created using a simple recommendation logic. 
The score increases when the content category matches the user's interest, when the content format matches the user's preferred format, when the activity type shows stronger engagement, and when the content is more popular.

For example, rows where both the interest and format match usually receive a higher score. 
Rows with no interest match receive a lower score, even if the content is popular.

#### Generating Personalized Recommendations

```python
content_recommendations = {
    "technology": "AI tutorial video",
    "music": "personalized music playlist",
    "fitness": "home workout tutorial",
    "gaming": "game strategy video",
    "travel": "travel guide blog",
    "education": "online learning tutorial",
    "movies": "movie recommendation list",
    "finance": "personal finance newsletter",
    "cooking": "quick recipe video",
    "fashion": "style inspiration blog"
}
```

```python
def generate_recommendation(row):
    return content_recommendations[row["interest"]]
```

```python
df_clean["generated_recommendation"] = df_clean.apply(generate_recommendation, axis=1)

df_clean[["interest", "preferred_format", "rule_based_score", "generated_recommendation"]].head()
```

A recommendation dictionary is used to match each user interest with a personalized content suggestion. 
The function generate_recommendation() uses the user's main interest to select the most relevant recommendation.

#### Recommendation Strength

```python
def classify_recommendation(score):
    if score >= 23:
        return "high"
    elif score >= 18:
        return "medium"
    else:
        return "low"

df_clean["recommendation_strength"] = df_clean["rule_based_score"].apply(classify_recommendation)

df_clean[["interest", "rule_based_score", "recommendation_strength", "generated_recommendation"]].head()
```

The numerical score is converted into a recommendation strength category. 
This makes the result easier to interpret by classifying recommendations as low, medium, or high.

The recommendation_strength column makes the rule-based score easier to interpret. 
A high score indicates that the content is a strong match for the user, while a low score means that the match is weaker.

The generated_recommendation column provides a personalized content suggestion based on the user's main interest.

Each row represents a user profile combined with a content interaction. 
Most users appear once in the dataset, while a few user IDs appear more than once, which can represent multiple interactions by the same user.

### Object-Oriented Programming Structure

This object-oriented section is a small way to imagine how the project could become a more complete program. 
The main analysis still happens with the DataFrame, but the classes help organize user information.

```python
class User:
    def __init__(self, name, age, interest, preferred_format):
        self.name = name
        self.age = age
        self._interest = interest
        self.preferred_format = preferred_format
        
    @property
    def interest(self):
        return self._interest
    
    def get_user_type(self):
        return "standard user"
```

```python
class PremiumUser(User):
    def __init__(self, name, age, interest, preferred_format, membership_level):
        super().__init__(name, age, interest, preferred_format)
        self.membership_level = membership_level
        
    def get_user_type(self):
        return "premium user"
```

```python
sample_user = User(
    name=df_clean.loc[0, "name"],
    age=df_clean.loc[0, "age"],
    interest=df_clean.loc[0, "interest"],
    preferred_format=df_clean.loc[0, "preferred_format"]
)
```

```python
standard_user = User("Demo User", 25, "technology", "video")
premium_user = PremiumUser("Premium Demo", 30, "music", "playlist", "gold")

print(standard_user.get_user_type())
print(premium_user.get_user_type())
```

```python
print("Sample user name:", sample_user.name)
print("Sample user interest:", sample_user.interest)
```

This section shows how user information could be organized with objects instead of only DataFrame rows.

The User class stores a basic profile, while PremiumUser represents a more specific type of user with an additional membership level. 
This simple structure gives an idea of how the project could later become a more complete recommendation program.

The main recommendation logic still comes from the DataFrame and the rule-based score.

The User class represents a simple user profile. 
It stores basic information such as name, age, interest, and preferred content format.

#### RecommendationEngine Class

The User class stores a simple user profile, while PremiumUser inherits from User and adds a membership level.

The _interest attribute is accessed through the interest property. 
Both User and PremiumUser have a get_user_type() method, but each class returns a different result.

```python
class RecommendationEngine:
    def __init__(self, recommendations):
        self.recommendations = recommendations
        
    def recommend(self, user):
        return self.recommendations[user.interest]
```

The main recommendation logic still comes from the DataFrame and the rule-based score. 
The object-oriented structure is used here as a simple way to organize user information and demonstrate how the project could be expanded into a more complete program.

```python
engine = RecommendationEngine(content_recommendations)

engine.recommend(sample_user)
```

The RecommendationEngine class stores the recommendation dictionary and uses a user's interest to generate a personalized content suggestion.

### Final Demo: User Recommendation Card

To make the recommendation system easier to understand, I created a final user card. 
It shows not only the recommendation, but also the information used to produce it.

This final demo selects one existing user from the dataset and displays the information used by the recommendation system. 
It shows the user's profile, interaction behavior, engagement signals, and final personalized recommendation.

```python
selected_user = df_clean.sample(1).iloc[0]

user_score = selected_user["rule_based_score"]
user_strength = selected_user["recommendation_strength"]
user_recommendation = selected_user["generated_recommendation"]
score_percentage = min((user_score / 30) * 100, 100)

base_score = 10
interest_bonus = 5 if selected_user["interest"] == selected_user["content_category"] else 0
format_bonus = 3 if selected_user["preferred_format"] == selected_user["content_format"] else 0
activity_bonus = activity_weights[selected_user["activity_type"]]
popularity_bonus = selected_user["content_popularity"] / 20

score_parts = {
    "Base score": base_score,
    "Interest match": interest_bonus,
    "Format match": format_bonus,
    "Activity signal": activity_bonus,
    "Popularity bonus": popularity_bonus
}

if user_strength == "high":
    match_message = "Excellent match based on this user's profile and activity."
elif user_strength == "medium":
    match_message = "Good match with some aligned preferences."
else:
    match_message = "Basic match. More user activity could improve the recommendation."
    
print("Personalized User Recommendation Card")
print("====================================")
print()
print("User Profile")
print("------------------------------------")
print("Name:", selected_user["name"])
print("Age:", selected_user["age"])
print("Main interest:", selected_user["interest"])
print("Preferred format:", selected_user["preferred_format"])
print()
print("Current Interaction")
print("------------------------------------")
print("Activity type:", selected_user["activity_type"])
print("Content category:", selected_user["content_category"])
print("Content format:", selected_user["content_format"])
print("Content popularity:", selected_user["content_popularity"])
print()
print("Behavior Signals")
print("------------------------------------")
print("Engagement score:", selected_user["engagement_score"])
print("Time spent:", selected_user["time_spent_minutes"], "minutes")
print("Clicks count:", selected_user["clicks_count"])
print("Previous interactions:", selected_user["previous_interactions"])
print("User rating:", selected_user["user_rating"])
print()
print()
print("=" * 50)
print("FINAL PERSONALIZED RECOMMENDATION")
print("=" * 50)
print("Recommended content:", user_recommendation.upper())
print("Recommendation strength:", user_strength.upper())
print("Match score:", round(score_percentage, 1), "%")
print(match_message)
print("=" * 50)

plt.figure(figsize=(8, 4))

plt.barh(
    list(score_parts.keys()),
    list(score_parts.values()),
    color=["gray", "seagreen", "mediumseagreen", "steelblue", "goldenrod"]
)

plt.title("Recommendation Score Breakdown")
plt.xlabel("Points Added to Recommendation Score")
plt.ylabel("Score Component")
plt.show()
```

It considers interest match, preferred format match, activity type, and content popularity to generate a recommendation score and a personalized content suggestion.

A limitation of this project is that the dataset is synthetic, so the results should be interpreted as a simulation rather than real user behavior. 
In the future, this system could be improved with real user data, more detailed user history, or a machine learning-based recommendation model.
